```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline
```

```python
df = pd.read_csv('D:/upgrad - data science/data sets/Uber Request Data.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])
```

```python
df.head()
```

```python
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'])
```

```python
df.head()
```

```python
plt.title('trip_status')
sns.countplot(df['Status'])
```

```python
df['Request Date'] = df['Request timestamp'].dt.date
df['Request Time'] = df['Request timestamp'].dt.time
```

```python
df.head()
```

```python
df['Drop Date'] = df['Drop timestamp'].dt.date
df['Drop Time'] = df['Drop timestamp'].dt.time
```

```python
df.head()
```

```python
df=df.drop(columns='Drop timestamp')
df=df.drop(columns='Request timestamp')
```

```python
df.head()
```

```python
df['days'] = df['Request Date'].apply(lambda x : dt.datetime.strftime(x,'%a'))
```

```python
df.head()
```

```python
qw = plt.subplots(figsize=(20,10))
plt.title('days_status')
qw = sns.countplot(x='days', hue='Status',data=df)
```

```python
df['hours'] = df['Request Time'].apply(lambda x:x.hour)
```

```python
df.head()
```

```python
qw = plt.subplots(figsize=(20,10))
plt.title('Hours')
qw = sns.countplot(x='hours',data=df)
```

```python
qw = plt.subplots(figsize=(20,10))
plt.title('locations')
qw = sns.countplot(x='hours',hue='Pickup point', data=df)
```

```python
city=df[df['Pickup point']=='City']
airport=df[df['Pickup point']=='Airport']
```

```python
city.head()
```

```python
qw = plt.subplots(figsize=(20,10))
plt.title('city to airport')
qw = sns.countplot(x='hours', hue='Status',data=city)
```

```python
qw = plt.subplots(figsize=(20,10))
plt.title('airport to city')
qw = sns.countplot(x='hours', hue='Status',data=airport)
```
